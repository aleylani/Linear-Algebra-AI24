# Solution to Matrix Equations

We are tasked with solving for $X$ in the equation $AX = B$, where $A$, $X$, and $B$ are all $2 \times 2$ matrices.

To solve for $X$, we need to first find the inverse of matrix $A$. 

Once we have $A^{-1}$, we can left-multiply both sides of the equation $AX = B$ by $A^{-1}$. This will yield $X = A^{-1}B$. 


---

### Question 1:

Solve for matrix $X$, given the equation $AX = B$, where:

$$
A = \begin{bmatrix} 
1 & 2 \\ 
3 & 4 
\end{bmatrix}, \quad
X = \begin{bmatrix} 
x_{11} & x_{12} \\ 
x_{21} & x_{22} 
\end{bmatrix}, \quad
B = \begin{bmatrix} 
5 & 6 \\ 
7 & 8 
\end{bmatrix}
$$

#### Step 1: Find the inverse of matrix $A$

The matrix $A$ is:

$$
A = \begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix}
$$

$ad-bc$ is:

$$
(1)(4) - (2)(3) = 4 - 6 = -2
$$

Thus, the inverse of $A$ is:

$$
A^{-1} = \frac{1}{-2} \begin{bmatrix} 4 & -2 \\ -3 & 1 \end{bmatrix}
= \begin{bmatrix} -2 & 1 \\ 1.5 & -0.5 \end{bmatrix}
$$

#### Step 2: Multiply $A^{-1}$ by matrix $B$

Now, multiply $A^{-1}$ by matrix $B$:

$$
B = \begin{bmatrix}
5 & 6 \\
7 & 8
\end{bmatrix}
$$

Performing the matrix multiplication $A^{-1}B$:

$$
X = \begin{bmatrix} -2 & 1 \\ 1.5 & -0.5 \end{bmatrix}
\begin{bmatrix}
5 & 6 \\
7 & 8
\end{bmatrix}
$$

The result is:

$$
X = \begin{bmatrix} (-2)(5) + (1)(7) & (-2)(6) + (1)(8) \\
(1.5)(5) + (-0.5)(7) & (1.5)(6) + (-0.5)(8) \end{bmatrix}
$$

$$
X = \begin{bmatrix} -10 + 7 & -12 + 8 \\
7.5 - 3.5 & 9 - 4 \end{bmatrix}
$$

$$
X = \begin{bmatrix} -3 & -4 \\
4 & 5 \end{bmatrix}
$$

So, the solution for $X$ is:

$$
X = \begin{bmatrix}
-3 & -4 \\
4 & 5
\end{bmatrix}
$$

#### Step 3: Verify the solution

We now verify that the matrix $X = \begin{bmatrix} -3 & -4 \\ 4 & 5 \end{bmatrix}$ satisfies the original equation $AX = B$:

$$
A = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}, \quad X = \begin{bmatrix} -3 & -4 \\ 4 & 5 \end{bmatrix}, \quad B = \begin{bmatrix} 5 & 6 \\ 7 & 8 \end{bmatrix}
$$

Multiplying $A$ by $X$:

$$
AX = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}
\begin{bmatrix} -3 & -4 \\ 4 & 5 \end{bmatrix}
= \begin{bmatrix} (1)(-3) + (2)(4) & (1)(-4) + (2)(5) \\
(3)(-3) + (4)(4) & (3)(-4) + (4)(5) \end{bmatrix}
$$

$$
AX = \begin{bmatrix} -3 + 8 & -4 + 10 \\
-9 + 16 & -12 + 20 \end{bmatrix}
= \begin{bmatrix} 5 & 6 \\
7 & 8 \end{bmatrix}
$$

Since $AX = B$, the solution is correct.

### Question 2:

Solve for matrix $X$, given the equation $AX = B$, where:

$$
A = \begin{bmatrix} 
2 & 1 \\
4 & 3 
\end{bmatrix}, \quad
X = \begin{bmatrix} 
x_{11} & x_{12} \\ 
x_{21} & x_{22} 
\end{bmatrix}, \quad
B = \begin{bmatrix} 
9 & 10 \\
11 & 12 
\end{bmatrix}
$$

#### Step 1: Find the inverse of matrix $A$

The matrix $A$ is:

$$
A = \begin{bmatrix}
2 & 1 \\
4 & 3
\end{bmatrix}
$$

$ad-bc$ is:

$$
(2)(3) - (1)(4) = 6 - 4 = 2
$$

Thus, the inverse of $A$ is:

$$
A^{-1} = \frac{1}{2} \begin{bmatrix} 3 & -1 \\ -4 & 2 \end{bmatrix}
= \begin{bmatrix} 1.5 & -0.5 \\ -2 & 1 \end{bmatrix}
$$

#### Step 2: Multiply $A^{-1}$ by matrix $B$

Now, multiply $A^{-1}$ by matrix $B$:

$$
B = \begin{bmatrix}
9 & 10 \\
11 & 12
\end{bmatrix}
$$

Performing the matrix multiplication $A^{-1}B$:

$$
X = \begin{bmatrix} 1.5 & -0.5 \\ -2 & 1 \end{bmatrix}
\begin{bmatrix}
9 & 10 \\
11 & 12
\end{bmatrix}
$$

The result is:

$$
X = \begin{bmatrix} (1.5)(9) + (-0.5)(11) & (1.5)(10) + (-0.5)(12) \\
(-2)(9) + (1)(11) & (-2)(10) + (1)(12) \end{bmatrix}
$$

$$
X = \begin{bmatrix} 13.5 - 5.5 & 15 - 6 \\
7.5 - 3.5 & 9 - 4 \end{bmatrix}
$$

$$
X = \begin{bmatrix} 8 & 9 \\
-7 & -8 \end{bmatrix}
$$

So, the solution for $X$ is:

$$
X = \begin{bmatrix}
8 & 9 \\
-7 & -8
\end{bmatrix}
$$

#### Step 3: Verify the solution

We now verify that the matrix $X = \begin{bmatrix} 8 & 9 \\ -7 & -8 \end{bmatrix}$ satisfies the original equation $AX = B$:

$$
A = \begin{bmatrix} 2 & 1 \\ 4 & 3 \end{bmatrix}, \quad X = \begin{bmatrix} 8 & 9 \\ -7 & -8 \end{bmatrix}, \quad B = \begin{bmatrix} 9 & 10 \\ 11 & 12 \end{bmatrix}
$$

Multiplying $A$ by $X$:

$$
AX = \begin{bmatrix} 2 & 1 \\ 4 & 3 \end{bmatrix}
\begin{bmatrix} 8 & 9 \\ -7 & -8 \end{bmatrix}
= \begin{bmatrix} (2)(8) + (1)(-7) & (2)(9) + (1)(-8) \\
(4)(8) + (3)(-7) & (4)(9) + (3)(-8) \end{bmatrix}
$$

$$
AX = \begin{bmatrix} 16 - 7 & 18 - 8 \\
32 - 21 & 36 - 24 \end{bmatrix}
= \begin{bmatrix} 9 & 10 \\
11 & 12 \end{bmatrix}
$$

Since $AX = B$, the solution is correct.

### Question 3:

Solve for matrix $X$ in the equation:

$$
A = \begin{bmatrix} 
3 & -1 \\
2 & 5 
\end{bmatrix}, \quad
X = \begin{bmatrix} 
x_{11} & x_{12} \\ 
x_{21} & x_{22} 
\end{bmatrix}, \quad
B = \begin{bmatrix} 
13 & 14 \\
15 & 16 
\end{bmatrix}
$$

#### Step 1: Find the inverse of matrix $A$

The matrix $A$ is:

$$
A = \begin{bmatrix}
3 & -1 \\
2 & 5
\end{bmatrix}
$$

$ad-bc$ is:

$$
(3)(5) - (-1)(2) = 15 + 2 = 17
$$

Thus, the inverse of $A$ is:

$$
A^{-1} = \frac{1}{17} \begin{bmatrix} 5 & 1 \\ -2 & 3 \end{bmatrix}
= \begin{bmatrix} \frac{5}{17} & \frac{1}{17} \\ -\frac{2}{17} & \frac{3}{17} \end{bmatrix}
$$

#### Step 2: Multiply $A^{-1}$ by matrix $B$

Now, multiply $A^{-1}$ by matrix $B$:

$$
B = \begin{bmatrix}
13 & 14 \\
15 & 16
\end{bmatrix}
$$

Performing the matrix multiplication $A^{-1}B$:

$$
X = \begin{bmatrix} \frac{5}{17} & \frac{1}{17} \\ -\frac{2}{17} & \frac{3}{17} \end{bmatrix}
\begin{bmatrix}
13 & 14 \\
15 & 16
\end{bmatrix}
$$

The result is:

$$
X = \begin{bmatrix} \left(\frac{5}{17}\right)(13) + \left(\frac{1}{17}\right)(15) & \left(\frac{5}{17}\right)(14) + \left(\frac{1}{17}\right)(16) \\
\left(-\frac{2}{17}\right)(13) + \left(\frac{3}{17}\right)(15) & \left(-\frac{2}{17}\right)(14) + \left(\frac{3}{17}\right)(16) \end{bmatrix}
$$

$$
X = \begin{bmatrix} \frac{65}{17} + \frac{15}{17} & \frac{70}{17} + \frac{16}{17} \\
\frac{-26}{17} + \frac{45}{17} & \frac{-28}{17} + \frac{48}{17} \end{bmatrix}
$$

$$
X = \begin{bmatrix} \frac{80}{17} & \frac{86}{17} \\
\frac{19}{17} & \frac{20}{17} \end{bmatrix}
$$

So, the solution for $X$ is:

$$
X = \begin{bmatrix}
\frac{80}{17} & \frac{86}{17} \\
\frac{19}{17} & \frac{20}{17}
\end{bmatrix}
$$

#### Step 3: Verify the solution

We now verify that the matrix $X = \begin{bmatrix} \frac{80}{17} & \frac{86}{17} \\ \frac{19}{17} & \frac{20}{17} \end{bmatrix}$ satisfies the original equation $AX = B$:

$$
A = \begin{bmatrix} 3 & -1 \\ 2 & 5 \end{bmatrix}, \quad X = \begin{bmatrix} \frac{80}{17} & \frac{86}{17} \\ \frac{19}{17} & \frac{20}{17} \end{bmatrix}, \quad B = \begin{bmatrix} 13 & 14 \\ 15 & 16 \end{bmatrix}
$$

Multiplying $A$ by $X$:

$$
AX = \begin{bmatrix} 3 & -1 \\ 2 & 5 \end{bmatrix}
\begin{bmatrix} \frac{80}{17} & \frac{86}{17} \\ \frac{19}{17} & \frac{20}{17} \end{bmatrix}
= \begin{bmatrix} (3)(\frac{80}{17}) + (-1)(\frac{19}{17}) & (3)(\frac{86}{17}) + (-1)(\frac{20}{17}) \\
(2)(\frac{80}{17}) + (5)(\frac{19}{17}) & (2)(\frac{86}{17}) + (5)(\frac{20}{17}) \end{bmatrix}
$$

$$
AX = \begin{bmatrix} \frac{240}{17} - \frac{19}{17} & \frac{258}{17} - \frac{20}{17} \\
\frac{160}{17} + \frac{95}{17} & \frac{172}{17} + \frac{100}{17} \end{bmatrix}
$$

$$
AX = \begin{bmatrix} \frac{221}{17} & \frac{238}{17} \\
\frac{255}{17} & \frac{272}{17} \end{bmatrix}
= \begin{bmatrix} 13 & 14 \\
15 & 16 \end{bmatrix}
$$

Since $AX = B$, the solution is correct.
