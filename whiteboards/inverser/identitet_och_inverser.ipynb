## Identity matrices and Inverse of Matrices


**Identity Matrices**

An **identity matrix** is a special matrix that plays a critical role in matrix operations. 

It is a square matrix with all diagonal entries equal to 1 and all off-diagonal entries equal to 0.

- The $2 \times 2$ identity matrix is:

$$I = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix}.$$

- The $3 \times 3$ identity matrix is:

$$I = \begin{pmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{pmatrix}.$$


*Important property of identity matrices*

The identity matrices has the property that, when multiplying it with any other matrix **A**, the result is the original matrix **A** itself. 

In other words:

$$AI = IA = A.$$

This is why these matrices are called *identity* matrices.



---

**Examples of Multiplication with Identity Matrices**

Let's prove the identity property by demonstration.

Assume we have $$A = \begin{pmatrix} a_{11} & a_{12} \\ a_{21} & a_{22} \end{pmatrix}$$:

Multiplying $A$ with the $2 \times 2$ identity matrix $I$:

$$A \cdot I = \begin{pmatrix} a_{11} & a_{12} \\ a_{21} & a_{22} \end{pmatrix} \cdot \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix} = \begin{pmatrix} a_{11} & a_{12} \\ a_{21} & a_{22} \end{pmatrix} = A.$$

Similarly:

$$I \cdot A = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix} \cdot \begin{pmatrix} a_{11} & a_{12} \\ a_{21} & a_{22} \end{pmatrix} = \begin{pmatrix} a_{11} & a_{12} \\ a_{21} & a_{22} \end{pmatrix} = A.$$

Anologously, assume we have $$B = \begin{pmatrix} b_{11} & b_{12} & b_{13} \\ b_{21} & b_{22} & b_{23} \\ b_{31} & b_{32} & b_{33} \end{pmatrix}$$:

Multiplying $B$ with the $3 \times 3$ identity matrix $I$:

$$B \cdot I = \begin{pmatrix} b_{11} & b_{12} & b_{13} \\ b_{21} & b_{22} & b_{23} \\ b_{31} & b_{32} & b_{33} \end{pmatrix} \cdot \begin{pmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{pmatrix} = \begin{pmatrix} b_{11} & b_{12} & b_{13} \\ b_{21} & b_{22} & b_{23} \\ b_{31} & b_{32} & b_{33} \end{pmatrix} = B.$$

Similarly:

$$I \cdot B = \begin{pmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{pmatrix} \cdot \begin{pmatrix} b_{11} & b_{12} & b_{13} \\ b_{21} & b_{22} & b_{23} \\ b_{31} & b_{32} & b_{33} \end{pmatrix} = \begin{pmatrix} b_{11} & b_{12} & b_{13} \\ b_{21} & b_{22} & b_{23} \\ b_{31} & b_{32} & b_{33} \end{pmatrix} = B.$$

**Inverse of Matrices**

The concept of the inverse of a matrix is fundamental in linear algebra. For a square matrix $A$, its inverse $A^{-1}$, if it exists, satisfies:

$$A A^{-1} = A^{-1} A = I,$$

where $I$ is the identity matrix. Not all matrices have an inverse though; a matrix must be for example be square to have an inverse.

Below, we will cover:
1. Finding the inverse of a $2 \times 2$ matrix using a known forula. 
2. Finding the inverse of a $3 \times 3$ matrix using the **Gauss-Jordan elimination method**.

For both, we will also prove that the resulting matrix is indeed the inverse by verifying $A A^{-1} = I$

---

**Inverse of a $2 \times 2$ Matrix**

Assume we have $$A = \begin{pmatrix} a & b \\ c & d \end{pmatrix}.$$

Its inverse $A^{-1}$ is then given by:

$$A^{-1} = \frac{1}{ad - bc} \begin{pmatrix} d & -b \\ -c & a \end{pmatrix}.$$

**Note** this is only valid if $ ad - bc \neq 0$, otherwise we say that $A$ is not invertible.

---

**Proof of the Inversion Property** 

$$A A^{-1} = I.$$

We have $$A = \begin{pmatrix} a & b \\ c & d \end{pmatrix},$$ and $$A^{-1} = \frac{1}{ad - bc} \begin{pmatrix} d & -b \\ -c & a \end{pmatrix}.$$ 

Compute the product $A A^{-1}$:

$$A \cdot A^{-1} = \begin{pmatrix} a & b \\ c & d \end{pmatrix} \cdot \frac{1}{ad-bc} \begin{pmatrix} d & -b \\ -c & a \end{pmatrix} = $$

$$\frac{1}{ad-bc}\begin{pmatrix} a & b \\ c & d \end{pmatrix} \cdot \begin{pmatrix} d & -b \\ -c & a \end{pmatrix} =$$
$$\frac{1}{ad-bc} \begin{pmatrix} ad - bc & -ab + ab \\ cd - cd & -bc + ad \end{pmatrix} =$$

$$\frac{1}{ad-bc} \begin{pmatrix} ad - bc & 0 \\ 0 & ad - bc \end{pmatrix} = $$

$$\begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix}.$$

Thus, $A A^{-1} = I$ is proven.

---

**Example: Finding the Inverse of a $2 \times 2$ Matrix**

Let:

$$A = \begin{pmatrix} 4 & 7 \\ 2 & 6 \end{pmatrix}.$$

**Step 1**: Calculate $ad - bc$ and verify it's non-zero.

$$
ad - bc = (4)(6) - (7)(2) = 24 - 14 = 10.
$$

Since $10 \neq 0$, we know that the inverse exists and can proceed.

**Step 2**: Apply the formula for the inverse

$$
A^{-1} = \frac{1}{10} \begin{pmatrix} 6 & -7 \\ -2 & 4 \end{pmatrix} =  \begin{pmatrix} \frac{6}{10} & \frac{-7}{10} \\ \frac{-2}{10} & \frac{4}{10} \end{pmatrix} = \begin{pmatrix} \frac{3}{5} & \frac{-7}{10} \\ \frac{-1}{5} & \frac{2}{5} \end{pmatrix}.
$$

Great. We've now found the inverse.

**Verification**:

To verify that that we've indeed found the inverse, we multiply $A$ by $A^{-1}$ and check if we get the identity matrix.

$$
A = \begin{pmatrix} 4 & 7 \\ 2 & 6 \end{pmatrix}, \quad A^{-1} = \begin{pmatrix} \frac{3}{5} & \frac{-7}{10} \\ \frac{-1}{5} & \frac{2}{5} \end{pmatrix}
$$

Performing the matrix multiplication $AA^{-1}$:

$$
AA^{-1} = \begin{pmatrix} 4 & 7 \\ 2 & 6 \end{pmatrix} \begin{pmatrix} \frac{3}{5} & \frac{-7}{10} \\ \frac{-1}{5} & \frac{2}{5} \end{pmatrix}
$$

Calculate each element of the resulting matrix:

1. First row, first column:
   $$
   4 \cdot \frac{3}{5} + 7 \cdot \frac{-1}{5} = \frac{12}{5} - \frac{7}{5} = \frac{5}{5} = 1.
   $$

2. First row, second column:
   $$
   4 \cdot \frac{-7}{10} + 7 \cdot \frac{2}{5} = \frac{-28}{10} + \frac{14}{5} = \frac{-28}{10} + \frac{28}{10} = 0.
   $$

3. Second row, first column:
   $$
   2 \cdot \frac{3}{5} + 6 \cdot \frac{-1}{5} = \frac{6}{5} - \frac{6}{5} = 0.
   $$

4. Second row, second column:
   $$
   2 \cdot \frac{-7}{10} + 6 \cdot \frac{2}{5} = \frac{-14}{10} + \frac{12}{5} = \frac{-14}{10} + \frac{24}{10} = \frac{10}{10} = 1.
   $$

Thus, the product is:

$$
A A^{-1} = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix}.
$$

This is the identity matrix, confirming that the inverse is correct.

**Final Answer**


The inverse of the matrix $A$ is:

$$
A^{-1} = \begin{pmatrix} \frac{3}{5} & \frac{-7}{10} \\ \frac{-1}{5} & \frac{2}{5} \end{pmatrix}.
$$

**Inverse of a $3 \times 3$ Matrix Using Gauss-Jordan Elimination**

Finding the inverse of a $3 \times 3$ matrix requires a little bit more work, since there is no closed-form formula for it.

Fortunately, it's still quite easy. We just use a trick.

Assume we have

$$A = \begin{pmatrix} 2 & 3 & 1 \\ 3 & 3 & 1 \\ 2 & 4 & 1 \end{pmatrix}.$$

To find $A^{-1}$, we first combine $A$ with the $3 \times 3$ identity matrix side by side

$$\begin{pmatrix} 
2 & 3 & 1 & \vert & 1 & 0 & 0 \\ 
3 & 3 & 1 & \vert & 0 & 1 & 0 \\ 
2 & 4 & 1 & \vert & 0 & 0 & 1 
\end{pmatrix}.$$

The goal is now, through row operations, transform the matrix on the left side into an identity matrix. Once that's done, the matrix on the right will be our inverse.

Let's demonstrate this.

**Perform Row Operations**

We now proceed with operations to achieve identity matrix on the left hand side.

1. Replace $R_2$ with $R_2 - R_1$:
   $$\begin{pmatrix} 
   2 & 3 & 1 & \vert & 1 & 0 & 0 \\ 
   1 & 0 & 0 & \vert & -1 & 1 & 0 \\ 
   2 & 4 & 1 & \vert & 0 & 0 & 1 
   \end{pmatrix}.$$

2. Swap $R_1$ and $R_2$:
   $$\begin{pmatrix} 
   1 & 0 & 0 & \vert & -1 & 1 & 0 \\ 
   2 & 3 & 1 & \vert & 1 & 0 & 0 \\ 
   2 & 4 & 1 & \vert & 0 & 0 & 1 
   \end{pmatrix}.$$

3. Replace $R_3$ with $R_3 - R_2$:
   $$\begin{pmatrix} 
   1 & 0 & 0 & \vert & -1 & 1 & 0 \\ 
   2 & 3 & 1 & \vert & 1 & 0 & 0 \\ 
   0 & 1 & 0 & \vert & -1 & 0 & 1 
   \end{pmatrix}.$$

4. Swap $R_2$ and $R_3$:
   $$\begin{pmatrix} 
   1 & 0 & 0 & \vert & -1 & 1 & 0 \\ 
   0 & 1 & 0 & \vert & -1 & 0 & 1 \\ 
   2 & 3 & 1 & \vert & 1 & 0 & 0 
   \end{pmatrix}.$$

5. Replace $R_3$ with $R_3 - 2R_1$:
   $$\begin{pmatrix} 
   1 & 0 & 0 & \vert & -1 & 1 & 0 \\ 
   0 & 1 & 0 & \vert & -1 & 0 & 1 \\ 
   0 & 3 & 1 & \vert & 3 & -2 & 0 
   \end{pmatrix}.$$

6. Replace $R_3$ with $R_3 - 3R_2$:
   $$\begin{pmatrix} 
   1 & 0 & 0 & \vert & -1 & 1 & 0 \\ 
   0 & 1 & 0 & \vert & -1 & 0 & 1 \\ 
   0 & 0 & 1 & \vert & 6 & -2 & -3 
   \end{pmatrix}.$$

We've now achieved the identity matrix on the left hand side!

**Extract the Inverse**

The right-hand side of the matrix is now the inverse of $A$:

$$A^{-1} = \begin{pmatrix} -1 & 1 & 0 \\ -1 & 0 & 1 \\ 6 & -2 & -3 \end{pmatrix}.$$
